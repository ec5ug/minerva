```python
import pandas as pd
import os
```

```python
os_dir = r"C:\Users\echan\PycharmProjects\minerva\data\school"
school_df = pd.DataFrame()
for file_name in os.listdir(os_dir):
    if file_name.endswith('.csv'):  # Assuming the files are CSV format
        file_path = os.path.join(os_dir, file_name)
        df = pd.read_csv(file_path, encoding='latin-1')  # Read the CSV file into a DataFrame
        school_df = pd.concat([school_df, df], ignore_index=1)
        
school_df = school_df[['LEANM05', 'SCHNAM05', 'LCITY05', 'LSTATE05', 'LATCOD', 'LONCOD']]
school_df.head(10)
```

```python
states = {
        'AK': 'Alaska',
        'AL': 'Alabama',
        'AR': 'Arkansas',
        'AS': 'American Samoa',
        'AZ': 'Arizona',
        'CA': 'California',
        'CO': 'Colorado',
        'CT': 'Connecticut',
        'DC': 'District of Columbia',
        'DE': 'Delaware',
        'FL': 'Florida',
        'GA': 'Georgia',
        'GU': 'Guam',
        'HI': 'Hawaii',
        'IA': 'Iowa',
        'ID': 'Idaho',
        'IL': 'Illinois',
        'IN': 'Indiana',
        'KS': 'Kansas',
        'KY': 'Kentucky',
        'LA': 'Louisiana',
        'MA': 'Massachusetts',
        'MD': 'Maryland',
        'ME': 'Maine',
        'MI': 'Michigan',
        'MN': 'Minnesota',
        'MO': 'Missouri',
        'MP': 'Northern Mariana Islands',
        'MS': 'Mississippi',
        'MT': 'Montana',
        'NA': 'National',
        'NC': 'North Carolina',
        'ND': 'North Dakota',
        'NE': 'Nebraska',
        'NH': 'New Hampshire',
        'NJ': 'New Jersey',
        'NM': 'New Mexico',
        'NV': 'Nevada',
        'NY': 'New York',
        'OH': 'Ohio',
        'OK': 'Oklahoma',
        'OR': 'Oregon',
        'PA': 'Pennsylvania',
        'PR': 'Puerto Rico',
        'RI': 'Rhode Island',
        'SC': 'South Carolina',
        'SD': 'South Dakota',
        'TN': 'Tennessee',
        'TX': 'Texas',
        'UT': 'Utah',
        'VA': 'Virginia',
        'VI': 'Virgin Islands',
        'VT': 'Vermont',
        'WA': 'Washington',
        'WI': 'Wisconsin',
        'WV': 'West Virginia',
        'WY': 'Wyoming'
}

def update_state(row):
    state = row['LSTATE05']
    if state in states.keys():
        return states[state]
    return state

school_df['LSTATE05'] = school_df.apply(update_state, axis=1)
```

```python
# Define a function to apply title casing to a string
def apply_title_case(s):
    if isinstance(s, str):
        return s.title()
    return s

school_df = school_df.applymap(apply_title_case)
school_df.head(10)
```

```python
school_df = school_df[~school_df['SCHNAM05'].str.contains('Preschool')]
school_df = school_df[~school_df['SCHNAM05'].str.contains('Pre-K')]

school_df = school_df[~school_df['SCHNAM05'].str.contains('Elementary')]
school_df = school_df[~school_df['SCHNAM05'].str.contains('Elem')]
school_df = school_df[~school_df['SCHNAM05'].str.contains('El')]
school_df = school_df[~school_df['SCHNAM05'].str.contains('Kindergarten')]
school_df = school_df[~school_df['SCHNAM05'].str.contains('Primary')]

school_df = school_df[~school_df['SCHNAM05'].str.contains('Middle')]

school_df = school_df[~school_df['SCHNAM05'].str.contains('Junior High')]
school_df = school_df[~school_df['SCHNAM05'].str.contains('J H')]
school_df = school_df[~school_df['SCHNAM05'].str.contains('7Th & 8Th')]
```

```python
school_df.sample(50)
```

```python
school_df = school_df.rename(columns={'LEANM05':'District', 'SCHNAM05':'School', 'LCITY05':'City', 'LSTATE05':'State', 
                               'LATCOD':'Latitude', 'LONCOD':'Longitude'})
school_df.head(10)
```
